### Tabla de frecuencia para variable **cualitativa** de un DataFrame de nombre **df**

```python
#Cálculo de frecuencia absoluta. La palabra frecuencia representa la frecuencia absoluta.
frecuencia = df.groupby('Nombre de la variable').size()

n = df.shape[0]

#Crear el DF con frecuencias absolutas y relativas
tabla = pd.DataFrame({
    'fi': frecuencia,
    'hi': round((frecuencia/n) *100, 1) # redondear según contexto y valores de los datos.
})
tabla
```

### Tabla de frecuencia para variable **cuantitativa continua** de un DataFrame de nombre **df**

```python
#Agregar una columna al DataFrame df

k = cantidad de intervalos #Debe ingresar este valor
n = tamaño de la muestra #Debe ingresar este valor
d = cantidad de decimales de los intervalos #Debe ingresar este valor

df['Intervalos'] = pd.cut(df['Nombre de la variable'], bins = k, include_lowest= True, precision = d)
frecuencia = df.groupby('Intervalos', observed = True).size()

tabla_continua = pd.DataFrame({
    'fi': frecuencia,
    'hi': (frecuencia/n)*100,
    'Fi':frecuencia.cumsum(),
    'Hi': ((frecuencia/n) * 100).cumsum()

})
tabla_continua #Puede indicar redondear con m cantidad de decimales round(tabla_continua,m)
```

### Gráfico para variable de un DataFrame de nombre **df**

### Medidas de posición y dispersión para una variable cuantitativa de un DataFrame de nombre **df**


```python
media = df['Nombre de la variable'].mean()

mediana = df['Nombre de la variable'].median()

moda = df['Nombre de la variable'].mode()

percentil_k = df['Nombre de la variable'].quantile(k) #ingrese el valor "k" en decimal

desviacion_estandar = df['Nombre de la variable'].std()
```

### Cálculo de medidas de asociación para el DataFrame **df**

```python
#Covarianza
covarianza = df['Variable independiente'].cov(df['Variable dependiente'])

#Coeficiente de correlación lineal de Pearson
coef_correlacion = df['Variable independiente'].corr(df['Variable dependiente'])
```

### Modelo de regresión lineal usando los datos del DataFrame **df**

```python
#Defición de variables independiente y dependiente
v_indep = df['Nombre de la variable independiente']
v_depen = df['Nombre de la variable dependiente']

modelo_lineal = np.polyfit(v_indep,v_depen,1)
pendiente, intercepto = modelo_lineal
```

### Coeficiente de determinación $R^{2}$ de un modelo de regresión lineal

```python
#Calcular predicciones según el modelo de regresión lineal obtenido
predicciones_v_depen = np.polyval(modelo_lineal, v_indep)

#Cálculo del coeficiente de determinación R2
R2 = r2_score(v_depen, predicciones_v_depen)
```

### Gráfico de dispersión

```python
#Defición de variables independiente y dependiente
v_indep = df['Nombre de la variable independiente']
v_depen = df['Nombre de la variable dependiente']

#Crear gráfico de dispersión
fig,ax = plt.subplots(figsize = (6,5))
dispersion = ax.scatter(v_indep, v_depen)

#Línea de tendencia
ax.plot(v_indep, pendiente * v_indep + intercepto, color='red', label= f'y={pendiente:.2f}x+{intercepto:.2f}')

#Personalización
ax.set_title('Título del gráfico')
ax.set_xlabel('Título eje x (unidades)')
ax.set_ylabel('Título eje y (unidades)')

ax.legend()
plt.show()
```
