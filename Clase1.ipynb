```python
import numpy as np
import pandas as pd

# --- Sección 1: Listas y Diccionarios de Notas ---
# 1. Lista de calificaciones
print("\n" + "*" * 30 + "\n")

notas = [6.5, 7.0, 4.3, 5.1, 2.5]
print("Lista de notas:", notas)
print("La primera calificación es:", notas[0])
print("\n" + "*" * 30 + "\n")
# 2. Diccionario de estudiantes y sus notas
print("\n" + "*" * 30 + "\n")
notas_estudiantes = {
    "Ana": 6.5,
    "Luis": 7.0,
    "Maria": 4.3,
    "Pedro": 5.1,
    "Pepe": 2.5
}
print("\nDiccionario de estudiantes:", notas_estudiantes)
print("\n" + "*" * 30 + "\n")
# 3. Cálculo del promedio de notas usando Numpy
print("\n" + "*" * 30 + "\n")
promedio_array = np.mean(list(notas_estudiantes.values()))
print(f"El promedio de las notas es: {promedio_array:.2f}")
print("\n" + "*" * 30 + "\n")
# --- Sección 2: Uso de Pandas DataFrame para Notas ---
# 4. Creación de un DataFrame a partir del diccionario de estudiantes
print("\n" + "*" * 30 + "\n")
df_notas = pd.DataFrame(list(notas_estudiantes.items()), columns=['Estudiante', 'Nota'])
print("\nDataFrame de Notas:")
print(df_notas)
print("\n" + "*" * 30 + "\n")

# Promedio del DataFrame
print("\n" + "*" * 30 + "\n")
promedio_df = df_notas["Nota"].mean()
print(f"\nEl promedio general es: {promedio_df:.2f}")
print("\n" + "*" * 30 + "\n")
# Mostrar estudiantes con notas menores a 5.0
print("\n" + "*" * 30 + "\n")
df_bajo = df_notas[df_notas["Nota"] < 5.0]
print("\nEstudiantes con nota menor a 5.0:")
print(df_bajo)
print("\n" + "*" * 30 + "\n")
# Encontrar el estudiante con la nota más alta
print("\n" + "*" * 30 + "\n")
idx_max = df_notas['Nota'].idxmax()
nombre_top = df_notas.loc[idx_max, 'Estudiante']
nota_top = df_notas.loc[idx_max, 'Nota']
print(f"\nEl estudiante con la nota más alta es {nombre_top} con un {nota_top}")
print("\n" + "*" * 30 + "\n")
```
